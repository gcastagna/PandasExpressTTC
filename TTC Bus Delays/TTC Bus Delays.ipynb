```python
import pandas as pd
```

This is the raw data provided by the city about TTC bus delays. It covers several years and looks to be a relatively clean dataset. Looking at one of the .xls files for now:

```python
# Load data for 2022
df_delay = pd.read_excel('ttc-bus-delay-data-2022.xlsx')
```

```python
# Look at the DataFrame

df_delay.info()

print("\nThis dataframe has "+str(len(df_delay))+ ' rows\n')

print('there are '+ str(len(set(df_delay['Incident']))) + ' unique types of delays in this dataframe\n')

print('The delay types are:\n '+ str(set(df_delay['Incident'])) )

df_delay.head()
```

I will now try to convert the Date and Time columns to a new Datetime column

```python
try:
    df_delay['Date'] = pd.to_datetime(df_delay['Date'])

    # Convert 'Time' column to Timedelta format
    df_delay['Time'] = pd.to_timedelta(df_delay['Time'] + ':00')

    # Combine 'Date' and 'Time' into a new column 'DateTime'
    df_delay['DateTime'] = df_delay['Date'] + df_delay['Time']
except:
    pass

# View dataframe
df_delay.head()
```
