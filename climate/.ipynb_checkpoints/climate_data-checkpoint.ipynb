```python
import pandas as pd
```

Daily data accessed from: https://climatedata.ca/download/#station-download

Hourly data accessed from: https://climate-change.canada.ca/climate-data/#/hourly-climate-data

Explore single daily date file:


```python
df_daily_weather = pd.read_csv('weather_daily.csv')
```

```python
df_daily_weather.info()

print("\nThis dataframe has "+str(len(df_daily_weather))+ ' rows\n')

print('there are '+ str(len(set(df_daily_weather['LOCAL_DATE']))) + ' unique days in this dataframe\n')

df_daily_weather.head()
```

Looks like this is the daily average or cumulative

Lets look at one of the hourly files:

```python
df_hourly_weather = pd.read_csv('climate-hourly (12).csv')
df_hourly_weather['UTC_DATE'] = pd.to_datetime(df_hourly_weather['UTC_DATE'])
```

```python
df_hourly_weather.info()

print("\nThis dataframe has "+str(len(df_hourly_weather))+ ' rows\n')

print('there are '+ str(len(set(df_hourly_weather['LOCAL_DAY']))) + ' unique days in this dataframe\n')

print('Data for this batch is available between '+ str(df_hourly_weather['UTC_DATE'].min()) + ' to '+ str(df_hourly_weather['UTC_DATE'].max()))

df_hourly_weather.head()
```
